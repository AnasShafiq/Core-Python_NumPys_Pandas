```python
#importing a CSV to Pandas
data = pd.read_csv('/content/sample_data/california_housing_test.csv')
data.head()
type(data)
```

```python
data.shape
data.info()
```

```python
# to find the value count of any column
data.value_counts('total_rooms')
```

```python
#grouping values by there mean
data.groupby('total_bedrooms').mean()
```

```python
#finding the stats of the dataframe
print(data.count())
print('--------------------------------------------------------------------------------')
print(data.mean())
print('--------------------------------------------------------------------------------')
print(data.std())
```

```python
# for all the statistics
data.describe()
```
